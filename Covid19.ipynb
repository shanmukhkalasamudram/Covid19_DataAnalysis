```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt 
print('Modules are imported.')
```

```python
confirm = pd.read_csv("Datasets/covid19_Confirmed_dataset.csv")
confirm
```

```python
confirm.drop(["Lat","Long"],axis=1,inplace = True)
```

```python
confirm.head(10)
```

```python
confirm_aggregated = confirm.groupby("Country/Region").sum()
```

```python
confirm_aggregated.head(10)
```

```python
confirm_aggregated.loc["India"].diff().plot()  #diff plot 
```

```python
confirm_aggregated.loc["India"].diff().max() # maximum for one country 
```

```python
countries = list(confirm_aggregated.index)
max_inf_rate = []
for c in countries:
    max_inf_rate.append(confirm_aggregated.loc[c].diff().max())
confirm_aggregated['max_inf_rate'] = max_inf_rate
```

```python
confirm_aggregated.head() # max inflection rate for each country 
```

```python
corona_data = pd.DataFrame(confirm_aggregated["max_inf_rate"])
```

```python
corona_data.head()  #only max inf rate 
```
